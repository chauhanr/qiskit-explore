```python
from qiskit import * 
from qiskit.tools.visualization import plot_bloch_multivector
from qiskit.tools.visualization import plot_histogram
import math 
```

```python
Aer.backends()
```

```python
qasm_simulator = Aer.get_backend('qasm_simulator')
statevector_simulator = Aer.get_backend('statevector_simulator')
```

```python
def run_on_qasm_simulator(circuit): 
    statevec_job = execute(circuit, backend=statevector_simulator)
    result = statevec_job.result()
    statevec = result.get_statevector()
    
    num_qubits = circuit.num_qubits
    circuit.measure([i for i in range(num_qubits)], [i for i in range(num_qubits)])  # assume qubit = classical bits
    qasm_job = execute(circuit, backend=qasm_simulator, shots=1024).result()
    counts = qasm_job.get_counts()
    
    return statevec, counts
```

```python
circuit = QuantumCircuit(2,2)
statevec, counts = run_on_qasm_simulator(circuit)
```

```python
plot_bloch_multivector(statevec)
```

```python
circuit.h(0)
statevec, counts = run_on_qasm_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
plot_histogram([counts])
```

```python
# making the complete circuit 
circuit = QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0,1)
statevec, counts = run_on_qasm_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
plot_histogram([counts])
```

```python
## Rotations with qubits 
circuit = QuantumCircuit(2,2)
circuit.rx(math.pi/4, 0)
circuit.rx(math.pi/2, 1)
statevec, counts = run_on_qasm_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
plot_histogram([counts])
```

```python
## Z gate
circuit = QuantumCircuit(1,1)
circuit.h(0)
circuit.z(0)
statevec, counts = run_on_qasm_simulator(circuit)
plot_bloch_multivector(statevec)
```

```python
plot_histogram([counts])
```
